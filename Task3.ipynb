```python
from typing import Type
from abc import ABC, abstractmethod

class Company(ABC):
    def __int__(self):
        self.total = 0


    IsInDepartment = None

    def doOperation(self):
        pass

    def GetTotalSalary(self, department):
        self.total = 0
        for component in department.employees:
            self.total = self.total + component.pay
        print (f"{department.name} Total salary: {self.total}")


class Employee(Company):
    def __init__(self, name, pay):
        self.name = name
        self.pay = pay

    def doOperation(self):
        return f"{self.name} Doing Operation"
    
    def RemoveEmployee(self):
        if self.IsInDepartment is not None:
            self.IsInDepartment.RemoveEmployee(self)
    
class Department(Company):
    def __init__(self, name):
        self.name = name
        self.employees = []

    def doOperation(self):
        return None

    def AddEmployee(self, component):
        component.RemoveEmployee()
        component.IsInDepartment = self
        self.employees.append(component)

    def RemoveEmployee(self, component):
        self.employees.remove(component)

    def MoveEmployee(self):
        "Detaching this folder from its parent folder"
        if self.IsInDepartment is not None:
            self.IsInDepartment.delete(self)
            self.IsInDepartment = None
            
    def Work(self, indent=""):
        print(f"{self.name}:          Number of Employees: {len(self.employees)}")
        for component in self.employees:
            print(component.doOperation())
```

```python
"A use case of the composite pattern."
Department1 = Department("Department 1")

Employee1 = Employee("Employee 1", 100)
Employee2 = Employee("Employee 2", 150)
Employee3 = Employee("Employee 3", 200)

Department1.AddEmployee(Employee1)
Department1.AddEmployee(Employee2)
Department1.AddEmployee(Employee3)

Department2 = Department("Department 2")

Employee4 = Employee("Employee 4", 250)
Department2.AddEmployee(Employee4)

Department3 = Department("Department 3")
Employee5 = Employee("Employee 5", 300)
Department3.AddEmployee(Employee5)
Employee6 = Employee("Employee 6", 350)
Department3.AddEmployee(Employee6)

Department1.Work()
Department2.Work()
Department3.Work()

C = Company()

C.GetTotalSalary(Department1)
C.GetTotalSalary(Department2)
C.GetTotalSalary(Department3)


```
